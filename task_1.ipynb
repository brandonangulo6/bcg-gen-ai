```python
import pandas as pd
```

```python
df = pd.read_csv('companies_10k_reports.csv', usecols=range(7))

for col in df.columns[2:]:
    df[col] = df[col].str.replace(',', '').astype(int)
```

```python
df = df.sort_values(by=['Company', 'Fiscal Year'])
df['Revenue Growth (%)'] = df.groupby(['Company'])['Total Revenue (in millions USD)'].pct_change() * 100
df['Net Income Growth (%)'] = df.groupby(['Company'])['Net Income (in millions USD)'].pct_change() * 100
df.head(10)
```

```python
summary_stats = df.groupby('Company')[[
    'Total Revenue (in millions USD)',
    'Net Income (in millions USD)',
    'Operating Cash Flow (in millions USD)',
    'Revenue Growth (%)',
    'Net Income Growth (%)'
]].agg(['mean', 'std', 'min', 'max'])

summary_stats.head()
```

```python
trend_by_year = df.groupby('Fiscal Year')[[
    'Total Revenue (in millions USD)',
    'Net Income (in millions USD)'
]].mean().reset_index()
trend_by_year.head()
```

## Summary of Financial Performance Trends

- **Apple** showed a slight revenue dip in 2023 followed by a modest recovery in 2024. However, its net income declined both years, suggesting increasing costs or other profitability challenges.
- **Microsoft** demonstrated strong, accelerating revenue growth, especially in 2024 (+15.7%). Net income growth also surged in 2024 (+21.8%), indicating operational efficiency or favorable margins.
- **Tesla** had strong revenue growth in 2023 (+18.8%) but saw a massive net income drop in 2024 (-52.2%), possibly due to high expenditures or reduced margins despite rising revenue.

### Key Takeaways:
- Microsoft stands out as the most consistently growing company in both revenue and income.
- Tesla’s 2024 income collapse needs further analysis—cost spikes or margin shrinkage could be drivers.
- Apple’s plateauing metrics suggest a maturing or saturated market presence.

These trends highlight the importance of not just top-line growth but bottom-line health, especially when analyzing long-term sustainability.
